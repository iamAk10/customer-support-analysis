## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df['Issue_reported at'] = pd.to_datetime(
    df['Issue_reported at'],
    dayfirst=True,
    errors='coerce'
)

df['issue_responded'] = pd.to_datetime(
    df['issue_responded'],
    dayfirst=True,
    errors='coerce'
)

df['response_minutes'] = (
    df['issue_responded']
    -
    df['Issue_reported at']
).dt.total_seconds()/60


print("Data Wrangling Done")
````

### What all manipulations have you done and insights you found?

Date columns were converted into datetime format to enable time-based analysis. A new feature named response_minutes was created by calculating the difference between issue response time and issue reported time. This helps analyze support efficiency and its relationship with customer satisfaction. Initial preprocessing also identified multiple missing values and mixed data types requiring cleaning in later stages.
