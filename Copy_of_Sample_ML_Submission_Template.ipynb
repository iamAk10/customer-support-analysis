## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Three hypothetical statements were formulated based on exploratory analysis to understand whether operational factors such as response time, support channels, and product categories significantly influence customer satisfaction and support efficiency. Statistical testing was performed to validate these assumptions and identify meaningful business relationships.


### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0) -
There is no significant relationship between response time and customer satisfaction score.

Alternate Hypothesis (H1) -
There is a significant relationship between response time and customer satisfaction score.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

temp=df[
['response_minutes',
'CSAT Score']
].dropna()

corr,p=pearsonr(
temp['response_minutes'],
temp['CSAT Score']
)

print(
"Correlation:",
corr
)

print(
"P Value:",
p
)
````

##### Which statistical test have you done to obtain P-Value?

Pearson Correlation Test

##### Why did you choose the specific statistical test?

Pearson correlation was selected because both variables are numerical and the objective is to determine whether a linear relationship exists between them.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0) - Customer support channels do not significantly affect customer satisfaction.

Alternate Hypothesis (H1) - Customer support channels significantly affect customer satisfaction.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

groups=[]

for i in df[
'channel_name'
].dropna().unique():

    groups.append(
        df[
        df['channel_name']==i
        ]['CSAT Score'].dropna()
    )

stat,p=f_oneway(
*groups
)

print(
"P Value:",
p
)
````

##### Which statistical test have you done to obtain P-Value?

One Way ANOVA

##### Why did you choose the specific statistical test?

ANOVA compares numerical customer satisfaction values across multiple support channel categories.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0) - Product categories do not significantly affect support response time.

Alternate Hypothesis (H1) - Product categories significantly affect support response time.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
groups=[]

for i in df[
'Product_category'
].dropna().unique():

    groups.append(
        df[
        df['Product_category']==i
        ]['response_minutes'].dropna()
    )

stat,p=f_oneway(
*groups
)

print(
"P Value:",
p
)
````

##### Which statistical test have you done to obtain P-Value?

One Way ANOVA

##### Why did you choose the specific statistical test?

ANOVA compares response times across multiple product category groups.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
numeric_columns = df.select_dtypes(
include=['int64','float64']
).columns

for col in numeric_columns:

    Q1=df[col].quantile(
    0.25
    )

    Q3=df[col].quantile(
    0.75
    )

    IQR=Q3-Q1

    lower=Q1-1.5*IQR

    upper=Q3+1.5*IQR

    df[col]=np.where(
        df[col]<lower,
        lower,
        np.where(
            df[col]>upper,
            upper,
            df[col]
        )
    )

print(
"Outlier Treatment Done"
)
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

IQR based capping technique was used to reduce the influence of extreme values while preserving most of the dataset distribution.

### 3. Categorical Encoding

````python
# Encode your categorical columns
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

categorical_columns=df.select_dtypes(
include='object'
).columns

for col in categorical_columns:

    df[col]=le.fit_transform(
    df[col].astype(str)
    )

print(
"Encoding Complete"
)
````
